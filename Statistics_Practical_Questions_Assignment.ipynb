````python
'''
 Q1) Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
     After generating the list, find the following:
 (i) Write a Python function to calculate the mean of a given list of numbers.
 Create a function to find the median of a list of numbers.
 (ii) Develop a program to compute the mode of a list of integers.
 (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
 (iv) Write a Python function to find the geometric mean of a list of positive numbers.
 (v) Create a program to calculate the harmonic mean of a list of values.
 (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
 (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers.   '''

#Ans)

import random
from statistics import mean, median, mode
from scipy.stats import gmean, hmean, trim_mean

# Step 1: Generate a list of 100 random integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Step 2 (i): Function to calculate the mean of a list
def calculate_mean(numbers):
    """Calculate the mean (average) of a list of numbers."""
    return sum(numbers) / len(numbers)

# Step 2 (i): Function to calculate the median of a list
def calculate_median(numbers):
    """Calculate the median of a list of numbers."""
    sorted_numbers = sorted(numbers)  # Sort the list for median calculation
    n = len(sorted_numbers)
    if n % 2 == 0:
        # If even, return the average of the two middle numbers
        return (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2
    else:
        # If odd, return the middle number
        return sorted_numbers[n//2]

# Step 2 (ii): Program to compute the mode of a list
def calculate_mode(numbers):
    """Calculate the mode of a list of numbers."""
    return mode(numbers)

# Step 2 (iii): Function to calculate the weighted mean of values with weights
def calculate_weighted_mean(values, weights):
    """Calculate the weighted mean of a list of values with corresponding weights."""
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weights = sum(weights)
    return weighted_sum / total_weights

# Step 2 (iv): Function to calculate the geometric mean of a list of positive numbers
def calculate_geometric_mean(numbers):
    """Calculate the geometric mean of a list of positive numbers."""
    return gmean(numbers)

# Step 2 (v): Function to calculate the harmonic mean of a list of numbers
def calculate_harmonic_mean(numbers):
    """Calculate the harmonic mean of a list of numbers."""
    return hmean(numbers)

# Step 2 (vi): Function to determine the midrange of a list (average of min and max)
def calculate_midrange(numbers):
    """Calculate the midrange (average of minimum and maximum) of a list."""
    return (min(numbers) + max(numbers)) / 2

# Step 2 (vii): Program to calculate the trimmed mean, excluding a percentage of outliers
def calculate_trimmed_mean(numbers, proportion_to_cut=0.05):
    """Calculate the trimmed mean of a list, excluding a percentage of outliers."""
    return trim_mean(numbers, proportiontocut=proportion_to_cut)

# Example usage:
print("Generated List:", int_list)
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean (example with equal weights):", calculate_weighted_mean(int_list, [1]*len(int_list)))
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (5% trim):", calculate_trimmed_mean(int_list, proportion_to_cut=0.05))


````

````python
# Q3)  Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

# Ans)
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize the discrete random variable.
        :param outcomes: List of outcomes (values) of the random variable
        :param probabilities: List of probabilities associated with each outcome
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("The length of outcomes and probabilities must be the same.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("The sum of probabilities must be equal to 1.")

        self.outcomes = np.array(outcomes)
        self.probabilities = np.array(probabilities)

    def expected_value(self):
        """
        Calculate and return the expected value (mean) of the random variable.
        :return: Expected value of the random variable
        """
        return np.sum(self.outcomes * self.probabilities)

    def variance(self):
        """
        Calculate and return the variance of the random variable.
        :return: Variance of the random variable
        """
        mean = self.expected_value()
        return np.sum(self.probabilities * (self.outcomes - mean) ** 2)

# Example usage:
outcomes = [1, 2, 3, 4, 5]            # Possible outcomes of the random variable
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]  # Corresponding probabilities of the outcomes

# Instantiate the DiscreteRandomVariable class
random_variable = DiscreteRandomVariable(outcomes, probabilities)

# Calculate and print expected value and variance
print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
````

````python
# Q4)  Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

# Ans)
import numpy as np

class DieSimulator:
    def __init__(self, sides=6):
        """
        Initialize the die simulator with a given number of sides (default is 6 for a fair die).
        """
        self.sides = sides
        self.outcomes = np.arange(1, sides + 1)
        self.probabilities = np.full(sides, 1 / sides)  # Equal probability for each side

    def roll_die(self, num_rolls):
        """
        Simulate rolling the die a specified number of times.
        :param num_rolls: Number of rolls
        :return: List of results from the rolls
        """
        return np.random.choice(self.outcomes, size=num_rolls, p=self.probabilities)

    def expected_value(self):
        """
        Calculate the expected value of the die outcomes.
        :return: Expected value of a single roll
        """
        return np.sum(self.outcomes * self.probabilities)

    def variance(self):
        """
        Calculate the variance of the die outcomes.
        :return: Variance of a single roll
        """
        mean = self.expected_value()
        return np.sum(self.probabilities * (self.outcomes - mean) ** 2)

# Example usage
die = DieSimulator()  # Create a fair six-sided die

# Simulate rolling the die 1000 times
num_rolls = 1000
rolls = die.roll_die(num_rolls)

# Calculate expected value and variance based on theoretical probabilities
theoretical_expected_value = die.expected_value()
theoretical_variance = die.variance()

print("Simulated Rolls:", rolls[:10], "...")  # Show first 10 rolls as a sample
print("Theoretical Expected Value:", theoretical_expected_value)
print("Theoretical Variance:", theoretical_variance)

# Calculate empirical mean and variance from the simulation
empirical_mean = np.mean(rolls)
empirical_variance = np.var(rolls, ddof=1)  # Use sample variance formula

print("Empirical Mean (from simulation):", empirical_mean)
print("Empirical Variance (from simulation):", empirical_variance)
````

````python
# Q5)  Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

# Ans) import numpy as np

def generate_samples(distribution, params, sample_size=1000):
    """
    Generate random samples from a specified probability distribution and calculate the mean and variance.

    :param distribution: The type of distribution ('binomial' or 'poisson').
    :param params: Parameters for the distribution:
                   - For 'binomial': params should be a tuple (n, p) where 'n' is the number of trials, 'p' is the probability of success.
                   - For 'poisson': params should be a single value 'lambda' which is the event rate.
    :param sample_size: Number of samples to generate (default is 1000).
    :return: A dictionary with 'samples', 'mean', and 'variance' of the generated samples.
    """
    # Generate samples based on the specified distribution
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n=n, p=p, size=sample_size)
    elif distribution == 'poisson':
        lam = params
        samples = np.random.poisson(lam=lam, size=sample_size)
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")

    # Calculate mean and variance of the samples
    mean = np.mean(samples)
    variance = np.var(samples, ddof=1)  # Sample variance

    return {
        "samples": samples,
        "mean": mean,
        "variance": variance
    }

# Example usage:

# Generate samples from a Binomial distribution with n=10 trials and p=0.5 probability of success
binomial_result = generate_samples('binomial', params=(10, 0.5))
print("Binomial Distribution:")
print("Mean:", binomial_result["mean"])
print("Variance:", binomial_result["variance"])

# Generate samples from a Poisson distribution with lambda=4 (event rate)
poisson_result = generate_samples('poisson', params=4)
print("\nPoisson Distribution:")
print("Mean:", poisson_result["mean"])
print("Variance:", poisson_result["variance"])
````

````python
# Q6)  Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

# Ans)
import numpy as np

def gaussian_statistics(mean, std_dev, sample_size=1000):
    """
    Generate random samples from a Gaussian distribution and compute their mean, variance, and standard deviation.

    :param mean: The mean (expected value) of the Gaussian distribution.
    :param std_dev: The standard deviation of the Gaussian distribution.
    :param sample_size: The number of samples to generate (default is 1000).
    :return: A dictionary with the generated samples, mean, variance, and standard deviation.
    """
    # Generate random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

    # Calculate statistics of the samples
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples, ddof=1)  # Sample variance
    sample_std_dev = np.std(samples, ddof=1)   # Sample standard deviation

    return {
        "samples": samples,
        "mean": sample_mean,
        "variance": sample_variance,
        "std_dev": sample_std_dev
    }

# Example usage
mean = 0       # Mean of the Gaussian distribution
std_dev = 1    # Standard deviation of the Gaussian distribution
sample_size = 1000  # Number of samples

# Get the statistics of the generated Gaussian samples
result = gaussian_statistics(mean, std_dev, sample_size)

print("Generated Gaussian Samples (first 10):", result["samples"][:10])
print("Sample Mean:", result["mean"])
print("Sample Variance:", result["variance"])
print("Sample Standard Deviation:", result["std_dev"])
````

````python
# Q8) Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

#Ans)
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point x.

    :param x: The value at which to evaluate the PDF.
    :param mean: The mean () of the normal distribution.
    :param std_dev: The standard deviation () of the normal distribution.
    :return: The PDF value at x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    return coefficient * math.exp(exponent)

# Example usage
mean = 0      # Mean of the distribution
std_dev = 1   # Standard deviation of the distribution
x = 1         # Point at which to evaluate the PDF

pdf_value = normal_pdf(x, mean, std_dev)
print("PDF value at x =", x, "is:", pdf_value)
````

````python
# Q9)  Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

# Ans)
import math

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution at a given point x.

    :param x: The value at which to evaluate the CDF (x >= 0).
    :param rate: The rate parameter () of the exponential distribution.
    :return: The CDF value at x.
    """
    if x < 0:
        return 0  # CDF is 0 for x < 0 in an exponential distribution
    return 1 - math.exp(-rate * x)

# Example usage
rate = 0.5  # Example rate parameter (lambda)
x = 2       # Point at which to evaluate the CDF

cdf_value = exponential_cdf(x, rate)
print("CDF value at x =", x, "is:", cdf_value)
````

````python
# Q10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

#Ans)
import math

def poisson_pmf(k, rate):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution for a given k.

    :param k: The number of occurrences (k >= 0).
    :param rate: The average rate () of occurrences (mean of the distribution).
    :return: The PMF value at k.
    """
    if k < 0:
        return 0  # PMF is 0 for negative values of k in a Poisson distribution
    return (rate ** k * math.exp(-rate)) / math.factorial(k)

# Example usage
rate = 3   # Example rate (lambda) or average number of occurrences
k = 2      # Number of occurrences for which to calculate the PMF

pmf_value = poisson_pmf(k, rate)
print("PMF value for k =", k, "is:", pmf_value)
````

````python
'''
Q11) A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.
 To generate the data use the following command:
 ```python
 import numpy as np
 # 50 purchases out of 1000 visitors
 old_layout = np.array([1] * 50 + [0] * 950)
 # 70 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)
 ```
 Apply z-test to find which layout is successful   '''

# Ans)
import numpy as np
import scipy.stats as stats

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Step 1: Calculate the number of successes and sample sizes
n_old = len(old_layout)
n_new = len(new_layout)

p_old = np.sum(old_layout) / n_old  # Conversion rate for old layout
p_new = np.sum(new_layout) / n_new  # Conversion rate for new layout

# Step 2: Calculate the pooled proportion
p_pool = (np.sum(old_layout) + np.sum(new_layout)) / (n_old + n_new)

# Step 3: Calculate the z-statistic
z_statistic = (p_new - p_old) / np.sqrt(p_pool * (1 - p_pool) * (1/n_old + 1/n_new))

# Step 4: Calculate the p-value
p_value = 1 - stats.norm.cdf(z_statistic)  # One-tailed test

# Output the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Step 5: Conclusion
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new layout leads to a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")

````

````python
'''
Q12)  A tutoring service claims that its program improves students' exam scores. A sample of students who
participated in the program was taken, and their scores before and after the program were recorded.
 Use the below code to generate samples of respective arrays of marks:
 ```python
 before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
 after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])
 ```
 Use z-test to find if the claims made by tutor are true or false.   '''

# Ans)
import numpy as np
import scipy.stats as stats

# Generate the samples
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Step 1: Calculate means and standard deviations
mean_before = np.mean(before_program)
mean_after = np.mean(after_program)
std_before = np.std(before_program, ddof=1)  # Sample standard deviation
std_after = np.std(after_program, ddof=1)    # Sample standard deviation

# Step 2: Calculate the sample sizes
n_before = len(before_program)
n_after = len(after_program)

# Step 3: Calculate the z-statistic
z_statistic = (mean_after - mean_before) / np.sqrt((std_before**2 / n_before) + (std_after**2 / n_after))

# Step 4: Calculate the p-value
p_value = 1 - stats.norm.cdf(z_statistic)  # One-tailed test

# Output the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Step 5: Conclusion
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program improves students' exam scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in scores.")
````

````python
'''
Q13)  A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They
conduct a study and record blood pressure measurements before and after administering the drug.
 Use the below code to generate samples of respective arrays of blood pressure:
 ```python
 before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
 after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])
 ```
 Implement z-test to find if the drug really works or not.'''

 #Ans)
import numpy as np
import scipy.stats as stats

# Generate the samples
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Step 1: Calculate means and standard deviations
mean_before = np.mean(before_drug)
mean_after = np.mean(after_drug)
std_before = np.std(before_drug, ddof=1)  # Sample standard deviation
std_after = np.std(after_drug, ddof=1)    # Sample standard deviation

# Step 2: Calculate the sample sizes
n_before = len(before_drug)
n_after = len(after_drug)

# Step 3: Calculate the z-statistic
z_statistic = (mean_after - mean_before) / np.sqrt((std_before**2 / n_before) + (std_after**2 / n_after))

# Step 4: Calculate the p-value
p_value = stats.norm.cdf(z_statistic)  # One-tailed test

# Output the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Step 5: Conclusion
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The drug is effective in reducing blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant reduction in blood pressure.")

````

````python
'''
Q15) A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.
 Use the following data:
 ```python
 layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
 layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]    '''

#Ans)
import numpy as np
import scipy.stats as stats

# Data for the click-through rates
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Function to perform A/B test analysis
def ab_test_analysis(layout_a, layout_b):
    # Calculate means and standard deviations
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    std_b = np.std(layout_b, ddof=1)  # Sample standard deviation

    # Sample sizes
    n_a = len(layout_a)
    n_b = len(layout_b)

    # Calculate the t-statistic
    t_statistic = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Calculate degrees of freedom
    df = n_a + n_b - 2

    # Calculate the p-value for a two-tailed test
    p_value = 2 * stats.t.sf(np.abs(t_statistic), df)  # Two-tailed test

    return t_statistic, df, p_value

# Perform the A/B test analysis
t_statistic, degrees_of_freedom, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Output the results
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-value:", p_value)

# Step 5: Conclusion
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
else:
    print("Fail to reject the null hypothesis: No significant difference in click-through rates between the two layouts.")
````

````python
'''
Q18) An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
significant difference between the average salaries of male and female employees.
 Use the below code to generate synthetic data:
 ```python
 # Generate synthetic salary data for male and female employees
 np.random.seed(0)  # For reproducibility
 male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
 female_salaries = np.random.normal(loc=55000, scale=9000, size=20)    '''

# Ans)
import numpy as np
import scipy.stats as stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Function to analyze salary differences
def analyze_salary_gap(male_salaries, female_salaries):
    # Calculate means and standard deviations
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    std_male = np.std(male_salaries, ddof=1)  # Sample standard deviation
    std_female = np.std(female_salaries, ddof=1)  # Sample standard deviation

    # Sample sizes
    n_male = len(male_salaries)
    n_female = len(female_salaries)

    # Calculate the t-statistic
    pooled_variance = ((std_male ** 2) / n_male) + ((std_female ** 2) / n_female)
    t_statistic = (mean_male - mean_female) / np.sqrt(pooled_variance)

    # Calculate degrees of freedom
    df = n_male + n_female - 2  # Total number of observations - 2

    # Calculate the p-value for a two-tailed test
    p_value = 2 * stats.t.sf(np.abs(t_statistic), df)  # Two-tailed test

    return t_statistic, df, p_value

# Perform the analysis
t_statistic, degrees_of_freedom, p_value = analyze_salary_gap(male_salaries, female_salaries)

# Output the results
print("Male Salaries:", male_salaries)
print("Female Salaries:", female_salaries)
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-value:", p_value)

# Step 4: Conclusion
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant gender-based salary gap.")
else:
    print("Fail to reject the null hypothesis: There is no significant gender-based salary gap.")
````

````python
'''
Q21)  A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.
 Use the below code to generate data:
 ```python
 np.random.seed(0)
 age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)
 voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)    '''

# Ans)
import numpy as np
import pandas as pd
import scipy.stats as stats

# Generate data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({
    'Age Group': age_groups,
    'Voter Preference': voter_preferences
})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2_statistic, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Output the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_statistic)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)

# Interpret the result
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")

````
